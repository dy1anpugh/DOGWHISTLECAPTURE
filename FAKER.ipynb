```python
import csv
import networkx as nx
import matplotlib.pyplot as plt
from netgraph import Graph
import numpy as np

# Replace 'file.csv' with your actual file path
myarray = 'ALLURLS.csv'

# Initialize an empty list to store the strings
strings_array = []

# Open the file
with open(myarray, 'r') as file:
    # Use csv.reader to read the file line by line
    reader = csv.reader(file)
    for row in reader:
        if row:  # Check if the row is not empty
            strings_array.append(row[0])  # Add the first column of each row to the array


clean_array=[]
no_repeats=[]
for x in strings_array:
    
    
        if ':' in x:
                clean_array.append(x.split(' on')[0])
        elif '(' in x:
                    clean_array.append(x.split('(')[0])
        elif '-' in x:
                    clean_array.append(x.split('-')[0])
# ALLURLGraph = nx.Graph()


    

ALLURLGraph = nx.Graph()
# Add nodes and edges (representing relationships)
for x in range((len(clean_array))-1):
        # if ':' in strings_array:
        if clean_array[x] not in clean_array[x+1]:
            ALLURLGraph.add_edge(clean_array[x],clean_array[x+1])
    
    # Draw the graph with matplotlib
# nx.draw(ALLURLGraph, with_labels=True, node_size=0, style=':', linewidths=0, edge_color='gray', arrows=True, node_color="white", font_size=8, alpha=1)
# connectionstyle='arc3,rad=0.2'

from pylab import rcParams
rcParams['figure.figsize'] = 20, 10
pos = nx.spring_layout(ALLURLGraph, scale=20, k=3/np.sqrt(ALLURLGraph.order()))
d = dict(ALLURLGraph.degree)
nx.draw(ALLURLGraph, pos, node_color='red', 
        with_labels=True, 
        nodelist=d, 
        style=':',
        edge_color='black',
        node_size=[d[k]*300 for k in d]
        )
plt.show()
    
```
